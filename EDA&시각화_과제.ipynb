### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
# Q2
df1 = df.isnull().sum().to_frame()
df1.rename(columns={0:'missing_num'}, inplace=True)
df1
df2 = df.isnull().sum().to_frame()/len(df)*100
df2.rename(columns={0:'missing_%'}, inplace=True)
df2
df_null = pd.concat([df1, df2], axis=1)
df_null = df_null[(df_null['missing_num'] > 0)]
df_null = df_null.sort_values(by='missing_num', ascending=False)
df_null
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
# Q3
plt.figure(figsize = (15,4))
#sns.barplot(y = 'missing_%', data = df_null)
df_null = df_null.reset_index()
sns.barplot(x = 'index', y = 'missing_%', data = df_null)
plt.xticks(rotation = 45)
plt.show()
```

- #### <font color='orange'> Q4. 결측치가 적은 beds 컬럼을 좀 더 자세히 살펴보겠습니다.

```python
# Q4
df[df["beds"].isnull()].head(5)
```
